# SQL Testing Portfolio

This SQL testing is to validate that the data between Back-End and Front-End is correct, in this testing I choose a music app scenario.

I execute this test using DBeaver with SQLite as an SQL database. In this testing, I used datasets that consist of many tables and each table has a relation across tables, namely as follows:

1. Track table
   
   
2. Artist table

   
3. Album table
   
   
4. Genre table

   
5. PlaylistTrack table

   
6. Playlist table

   

#### 1. User Select a Particular Song

```sql
SELECT
	Table1.Name AS Song,
	Table3.Name AS Artist,
	Table2.Title AS Album
FROM
	Track AS Table1
INNER JOIN 
	Album AS Table2
ON (Table2.AlbumId = Table1.AlbumId)
INNER JOIN
	Artist AS Table3
ON (Table2.ArtistId = Table3.ArtistId)
WHERE Table1.TrackId = 1143
```

#### 2. User open the songs that grouped by genre

Before that, let's take a look at the list of genres that users can choose



```sql
SELECT 
	Table1.Name AS Genre, 
	Table2.Name AS Song, 
	Table3.Title AS Artist
FROM
	Genre AS Table1
INNER JOIN
	Track AS Table2
ON (Table2.GenreId = Table1.GenreId)
INNER JOIN
	Album AS Table3
ON (Table2.AlbumId = Table3.AlbumId)
INNER JOIN
	Artist AS Table4
ON (Table3.ArtistId = Table4.ArtistId)
WHERE Table1.Name = "Rock"
```

#### 3. User open a playlist that consists of songs

Before that, let's take a look at the list of song playlists that users can choose



```sql
SELECT 
	Table1.Name AS Playlist, 
	Table3.Name AS Song, 
	Table5.Name AS Artist
FROM
	Playlist AS Table1
INNER JOIN
	PlaylistTrack AS Table2
ON (Table2.PlaylistId = Table1.PlaylistId)
INNER JOIN
	Track AS Table3
ON (Table3.TrackId = Table2.TrackId)
INNER JOIN
	Album AS Table4
ON (Table4.AlbumId = Table3.AlbumId)
INNER JOIN
	Artist AS Table5
ON (Table5.ArtistId = Table4.ArtistId)
WHERE
	Table1.PlaylistId = 5
```
